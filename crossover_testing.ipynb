```python
import numpy as np
import itertools, random
from numpy.random import randint
np.set_printoptions(linewidth = 200)
import numpy as np
```

```python
x = np.zeros([16, 16])

for i in xrange(1000):
    i0 = randint(0, x.shape[0])
    j0 = randint(0, x.shape[1])
    
    i1 = randint(0, x.shape[0])
    j1 = randint(0, x.shape[1])

    for i in xrange(min(i0, i1), max(i0,i1)+1):
        for j in xrange(min(j0, j1), max(j0, j1)+1):
            x[i, j] += 1
            
print x
```

```python
x = np.zeros([16, 16])

for i in xrange(1000):
    i0, j0 = randint(0, x.shape[0]), randint(0, x.shape[1])
    h, w = randint(0, x.shape[0]), randint(0, x.shape[1])

    for i in xrange(i0, i0 + h):
        for j in xrange(j0, j0+w):
            x[i%x.shape[0], j%x.shape[1]] += 1
            
print x
```
